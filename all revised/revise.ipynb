# normilization 

```python
from sklearn .preprocessing import MinMaxScaler
```

# ordinal encoder 

```python
data=sns.load_dataset("titanic")
```

```python
data.head()
```

```python
data["class"].value_counts()
```

```python
from sklearn .preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[["Third","Second","First"]],dtype=int)
```

```python
data["class_oe"]=oe.fit_transform(data[["class"]])
```
