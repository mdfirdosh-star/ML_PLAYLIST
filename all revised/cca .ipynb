```python
new_col=data["Fare"].dropna()
```

```python
data.shape,  new_col.shape
```

```python
plt.figure(figsize=(16,4))
plt.subplot(121)
sns.distplot(data["Fare"])
plt.legend()


plt.subplot(122)
sns.distplot(new_col)
plt.legend()
```

```python
new_data=data["Age"].dropna()
```

```python
plt.figure(figsize=(16,4))
plt.subplot(121)
sns.kdeplot(data["Age"],label="Before cca",color="m")
plt.legend()


plt.subplot(122)
sns.kdeplot(new_data,label="after cca")
plt.legend()
```
