```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn .model_selection import train_test_split
from sklearn .preprocessing import OneHotEncoder,OrdinalEncoder,StandardScaler,MinMaxScaler,LabelEncoder
from sklearn .impute import SimpleImputer
from sklearn .compose import ColumnTransformer
from sklearn .pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
import pickle 
```

```python
x=data.drop(columns=['test preparation course'])
y=data["test preparation course"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
x_train["race/ethnicity"].value_counts()
```

```python
x_train["parental level of education"].value_counts()
```

```python
l=LabelEncoder()
```

```python
y_train=l.fit_transform(y_train)
y_test=l.transform(y_test)
```

```python
x_train.shape,y_train.shape
```

# use columns transformer

```python
step_1=ColumnTransformer([

("ordinalencoder",OrdinalEncoder(categories=[["group E","group D","group C","group B","group A"]]),[1]),
("Ordinal_2",OrdinalEncoder(categories=[["high school","some high school","some college","associate's degree","bachelor's degree","master's degree"]]),[2]),
("onehotencoder_1",OneHotEncoder(sparse_output=False,handle_unknown="ignore"),[0]),
 ("onehotencoder_2",OneHotEncoder(sparse_output=False,handle_unknown="ignore"),[3])],remainder="passthrough")
```

```python
step_2=ColumnTransformer([
    ("standardscaler",StandardScaler(),slice(0,11))
])
```

```python
step_3= LogisticRegression()
```

```python
model=Pipeline([
    ("setp_1",step_1),
     ("setp_2",step_2),
     ("setp_3",step_3)
    
])
```

```python
file="logesticRegression_2.pkl"
```

```python
model.fit(x_train,y_train)
```

```python
pickle.dump(model,open(file,"wb"))
```

```python
pred=model.predict(x_test)
```

```python
from sklearn .metrics import accuracy_score
```

```python
score=accuracy_score(pred,y_test)
```
