# topic: Feature Contractions and feature splitting 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn .linear_model import LogisticRegression

```

```python
data=pd.read_csv(r"C:\Users\mdfir\Downloads\train.csv")[["Survived","Pclass","SibSp","Parch","Age"]]
```

```python
data.head()
```

```python
data=data.dropna()
```

```python
data
```

```python
x=data.iloc[:,1:5]
y=data["Survived"]
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring="accuracy",cv=10))
```

# Appling the feature contractions 

```python
x["family_size"]=x["SibSp"]+x["Parch"]+1
```

```python
x
```

```python
# family devied in category 

def family_cal(num):
    # alone 
    if num==1:
        return 0
    # small family 
    elif num >1 and num <= 4:
        return 1
    # big family
    else:
        return 2
```

```python
x["family_type"]=x["family_size"].apply(family_cal)
```

```python
x.head()
```

# drop the columns 

```python
x.drop(columns=["SibSp","Parch","family_size"],inplace=True)
```

```python
x.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring="accuracy",cv=20))
```
