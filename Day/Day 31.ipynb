```python
import pandas as pd 
import numpy  as np
import matplotlib .pyplot as plt
import scipy.stats as stats
from sklearn .model_selection import train_test_split,cross_val_score
from sklearn .preprocessing import PowerTransformer
from sklearn .linear_model import LogisticRegression,LinearRegression
from sklearn.metrics import r2_score
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\mdfir\Downloads\concrete_data.csv")
```

```python
x=data.iloc[:,0:8]
y=data.iloc[:,-1]
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict (x_test)
```

```python
score= r2_score(y_test,pred)
```

```python
score
```

```python
cvs=cross_val_score(model,x,y,scoring="r2")
```

```python
cvs.mean()
```

# all columns distributions 

```python
for i in x_train.columns:
   plt.figure(figsize=(14,4))
   plt.subplot(121)
   sns.distplot(x_train[i])
   plt.title(i)

   plt.subplot(122)
   stats.probplot(x_train[i],dist="norm",plot=plt)
   plt.title(i)
   plt.show()
```

```python
pt=PowerTransformer(method="box-cox")
x_train_pt=pt.fit_transform(x_train+0.0000001)
x_test_pt=pt.transform(x_test+0.0000001)
pd.DataFrame({'cols':x_train.columns,"box_cox_lambdas":pt.lambdas_})
```

# use box_cox transformations 

```python
lr=LinearRegression()
```

```python
lr.fit(x_train_pt,y_train)
```

```python
y_pred=lr.predict(x_test_pt)
```

```python
score=r2_score(y_test,y_pred)
```

```python
score
```

```python
x_train_pt=pd.DataFrame(x_train_pt,columns=x_train.columns)
for i in x_train_pt.columns:
   plt.figure(figsize=(14,4))
   plt.subplot(121)
   sns.distplot(x_train_pt[i])
   plt.title(i)
   plt.title("afet the power transform")

   plt.subplot(122)
   sns.distplot(x_train[i])
   plt.title(i)
   plt.title("Before the power transform")
   plt.show()
   
```
