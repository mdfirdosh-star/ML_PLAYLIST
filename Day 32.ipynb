# use discretized encoding 

```python
kbin_age=KBinsDiscretizer(n_bins=5,encode="ordinal",strategy="uniform")
kbin_fare=KBinsDiscretizer(n_bins=5,encode="ordinal",strategy="uniform")
```

```python
cl=ColumnTransformer([
    ("first",kbin_age,[0]),
    ("second",kbin_fare,[1])
])
```

```python
x_train_cl=cl.fit_transform(x_train)
x_test_cl=cl.fit_transform(x_test)
```

# show the my bins columns 

```python
cl.named_transformers_["first"].n_bins_
```

```python
cl.named_transformers_["first"].bin_edges_
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(x_train_cl,y_train)
```

```python
y_pred=clf.predict(x_test_cl)
```

```python
y_pred
```

```python
from sklearn .metrics import accuracy_score
```

```python
score=np.mean(accuracy_score(y_test,y_pred))
```
